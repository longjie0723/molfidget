```python
import lib3mf
from lib3mf import *


filename = "output/cube_scene.3mf"
wrapper = Wrapper()
model = wrapper.CreateModel()
reader = model.QueryReader("3mf")
print(f"Reading {filename}...")
mesh = reader.ReadFromFile(filename)
```

```python
import trimesh

# Create a cube with trimesh
cube = trimesh.creation.box(extents=(1, 1, 1))
cube.visual.vertex_colors = [255, 0, 0, 255]  # Set color to red
# export to 3mf
cube.export('output/cube.3mf')

scene = trimesh.Scene()
scene.add_geometry(cube)
# Save the scene to a 3mf file
scene.export('output/cube_scene.3mf')
```
